```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Prepare training data
data = {
    'study_hours': [1, 2, 3, 4, 5, 6],
    'grade':       [5, 4, 3, 2, 2, 1]
}
df = pd.DataFrame(data)

plt.scatter(df['study_hours'], df['grade'], color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.title("Study Hours vs. Grade")
plt.grid(True)
plt.show()
```

```python
#Model training
X = df[['study_hours']]
y = df['grade']
model = LinearRegression()
model.fit(X, y)

new_data = pd.DataFrame({'study_hours': [3.5]})
predicted_grade = model.predict(new_data)
print(f"Predicted grade for 3.5 study hours: {predicted_grade[0]:.2f}")
```

```python
plt.scatter(X, y, label="Actual", color='blue')
plt.plot(X, model.predict(X), label="Model", color='red')
plt.scatter(3.5, predicted_grade, color='green', label="Prediction", zorder=5)
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.title("Linear Regression Prediction")
plt.legend()
plt.grid(True)
plt.show()
```
